```python
%matplotlib inline

import numpy as np
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import matplotlib.pyplot as plt
from sklearn.preprocessing import normalize
import scipy.optimize as op
```

```python
def sigmoid(z):
    return 1.0 / (1.0 + np.exp(-z))

def compute_cost(theta, X, y):
    h = sigmoid(X.dot(theta))
    cost = 1 / m * (-y.T.dot(np.log(h)) - (1 - y).T.dot(np.log(1 - h)))
    grad = 1 / m * X.T.dot(h - y)
    return (cost[0][0], grad)

def gradient_descent(X, y, theta, alpha=0.01, iterations=1500):
    J_hist = []
    for _ in range(iterations):
        h = sigmoid(X.dot(theta))
        theta = theta - alpha / m * X.T.dot(h - y)
        J_hist.append(compute_cost(theta, X, y))
    return (theta, J_hist)

def visualize_theta(positive, negative, theta=None):
    plt.plot(positive[:,0], positive[:,1], '+')
    plt.plot(negative[:,0], negative[:,1], 'o')
    
    if theta is not None:
        X = np.linspace(0.2, 1, 50)
        Y = - (theta[0] + theta[1] * X) / theta[2]
        plt.plot(X, Y, 'g-')
    
    plt.title('Training data' + (' with decision boundary' if theta is not None else ''))
    plt.show()
```

```python
data = np.genfromtxt('./ex2/ex2data1.txt', delimiter=',')
original_data = data
data = normalize(data, axis=0, norm='max')

original_data[:5]
```

```python
m = len(data)
n = len(data[0])

ones = np.ones((m, 1))
X = np.column_stack((ones, data[:,:2]))
y = data[:,2].reshape((m, 1))

positive = data[data[:,2] == 1]
negative = data[data[:,2] == 0]

visualize_theta(positive, negative)
```

```python
theta = np.zeros((n, 1))

print('Expected cost (approx): 0.693\n');
print('Expected gradients (approx):\n -0.1000\n -12.0092\n -11.2628\n');

theta_test_1 = theta
cost, grad = compute_cost(theta_test_1, X, y)
print('cost: ', cost)
print('grad: \n', grad)
```

## Using gradient descent

```python
# https://stackoverflow.com/questions/18801002/fminunc-alternate-in-numpy/21446032
# http://aimotion.blogspot.com/2011/11/machine-learning-with-python-logistic.html

theta_gradiend_descent, J_hist = gradient_descent(X, y, theta, alpha=0.2, iterations=20000);
print('Expected theta values (approx)\n');
print(' -25.161\n 0.206\n 0.201\n');

print('actual theta after gradient descent: ')
print(theta_gradiend_descent)

J_hist = list(map(lambda x: x[0], J_hist))

plt.plot(np.arange(0, len(J_hist)), J_hist, '-')
plt.title('Cost function')
plt.show()

visualize_theta(positive, negative, theta=theta_gradiend_descent)
```

## Using optimization algorithm 

```python
# https://stackoverflow.com/questions/18801002/fminunc-alternate-in-numpy/21446032
# http://aimotion.blogspot.com/2011/11/machine-learning-with-python-logistic.html

def gradient_descent_step(theta, X, y):
    theta = theta.reshape(n, 1)
    h = sigmoid(X.dot(theta))
    theta = X.T.dot(h - y) / m
    return theta.reshape(n)

def compute_cost(theta, X, y):
    theta = theta.reshape(n, 1)
    h = sigmoid(X.dot(theta))
    cost = 1 / m * (-y.T.dot(np.log(h)) - (1 - y).T.dot(np.log(1 - h)))
    return cost[0][0]

result = op.minimize(fun = compute_cost,
                     x0 = np.zeros(n),
                     args = (X, y),
                     method = 'TNC',
                     jac = gradient_descent_step)

theta_op = result.x

print('Theta after optimization alg: ', theta_op.T)

visualize_theta(positive, negative, theta=theta_op)
```

## Regularized logistic regression

```python
def map_features(x1, x2, degree=6):
    ones = np.ones((len(x1), 1))
    columns = [ones]

    for i in range(1, degree + 1):
        for j in range(0, i + 1):
            columns.append(x1 ** (i - j) * x2 ** j)
    
    return columns

def map_features_2(x1, x2, degree=6):
    columns = [1]
    for i in range(1, degree + 1):
        for j in range(0, i + 1):
            columns.append(x1 ** (i - j) * x2 ** j)
    return columns

def compute_cost_reg(theta, X, y, lambdaa=0, reg=True):
    h = sigmoid(X.dot(theta))
    m = len(y)
    
    cost = 1 / m * (-y.T.dot(np.log(h)) - (1 - y).T.dot(np.log(1 - h)))
    cost = cost[0]
    if reg:
        theta_reg = theta[1:]
        cost_reg = lambdaa / 2 / m * (theta_reg.T.dot(theta_reg))
        cost += cost_reg
    return cost

def gradient_descent_step(theta, X, y):
    n = len(theta)
    theta = theta.reshape(n, 1)
    h = sigmoid(X.dot(theta))
    theta = X.T.dot(h - y) / len(X)
    return theta.reshape(n)

# https://stackoverflow.com/questions/22294241/plotting-a-decision-boundary-separating-2-classes-using-matplotlibs-pyplot
def visualize_theta_contour(positive, negative, theta=None):
    plt.plot(positive[:,0], positive[:,1], '+')
    plt.plot(negative[:,0], negative[:,1], 'o')
    
    if theta is not None:
        U = np.linspace(-1, 1.5, 100)
        V = np.linspace(-1, 1.5, 100)
        
        Z = np.zeros((len(U), len(V)))
        for i in range(len(U)):
            for j in range(len(V)):
                Z[i,j] = np.array(map_features_2(U[i], V[j])).dot(theta)

        plt.contour(U, V, Z.T > 1, levels=[0.5])
    
    plt.title('Training data' + (' with decision boundary' if theta is not None else ''))
    plt.show()
    
def visualize_cost_function(J_hist):
    plt.plot(np.arange(0, len(J_hist)), J_hist, 'g-')
    plt.title('Cost function')
    plt.show()

# create_compute_cost(1)(np.ones(len(X[0])), X, y)
```

```python
data = np.genfromtxt('./ex2/ex2data2.txt', delimiter=',')
original_data = data
# data = normalize(data, axis=0, norm='max')

m = len(data)

positive = data[data[:,2] == 1]
negative = data[data[:,2] == 0]

X1 = data[:,0]
X2 = data[:,1]

X = np.column_stack(map_features(X1, X2))
y = data[:,2].reshape((m, 1))

visualize_theta_contour(positive, negative)
```

```python
theta_test = np.ones(len(X[0]))
cost = compute_cost_reg(theta_test, X, y, lambdaa=10)

print('Expected cost (approx): 3.16')
print('Cost at test theta (with lambda = 10): ', cost)
```

```python
J_hist = []

def create_compute_cost(lambdaa=0, reg=True):
    def compute_cost_internal(theta, X, y):
        cost = compute_cost_reg(theta, X, y, lambdaa, reg)
        J_hist.append(cost)
        return cost
    
    return compute_cost_internal

lambdaa = 0.005
result = op.fmin_cg(f = create_compute_cost(lambdaa, reg=True),
                    x0 = np.zeros(len(X[0])),
                    args = (X, y),
                    fprime = gradient_descent_step)

theta_op = result

visualize_cost_function(J_hist)
visualize_theta_contour(positive, negative, theta=theta_op)
# print(theta_op)
```
